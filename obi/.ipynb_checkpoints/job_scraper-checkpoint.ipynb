```python
#Parsing functions from Rick
def parse_headers(headers):
    return dict([i.strip().split(': ') for i in headers.strip().split('\n')])

raw_headers = """Accept: text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3
Accept-Encoding: gzip, deflate, br
Accept-Language: en-US,en;q=0.9,vi;q=0.8
Connection: keep-alive
Host: www.jobmarketinsights.com
Referer: https://www.jobmarketinsights.com/jmi-area/Landing
Sec-Fetch-Mode: navigate
Sec-Fetch-Site: same-origin
Sec-Fetch-User: ?1
Upgrade-Insecure-Requests: 1
User-Agent: Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"""

headers = parse_headers(raw_headers)
```

```python
#Helper Functions from Rick's code
def throttle_request(min_sleep=0.5, max_sleep=1.0):
    time.sleep(random.uniform(min_sleep, max_sleep))

def simulate_typing(text, element, min_sleep=0.05, max_sleep=0.2):
    for c in text:
        element.send_keys(c)
        throttle_request(min_sleep, max_sleep)

def parse_cookies(cookies, cookie_sort=None):
    return '; '.join([
        (cookie['name'] + '=' + cookie['value'])
        for cookie in cookies
    ])

def login_to_jmi():
    driver.get('https://www.jobmarketinsights.com/jmi-area/Login')
    if driver.current_url == 'https://www.jobmarketinsights.com/jmi-area/Login':
        username = driver.find_element_by_id("email")
        password = driver.find_element_by_id("password")

        simulate_typing('patrickg@kalibrr.com', username)
        simulate_typing('$lGghdzv%B9id0t%do#RR7l5RsLW7lPmJTCuF$XhX049eyfnR', password)
        driver.find_element_by_xpath("//input[@type='submit']").click()
    elif driver.current_url == 'https://www.jobmarketinsights.com/jmi-area/Summary':
        print('pass')
        pass
    else:
        raise ValueError(f'Forwarded to an unknown url; {driver.current_url}')
    
    cookies = driver.get_cookies()
    #cookies[1]['value'] = '-1'
    return parse_cookies(cookies)
```

```python
#URL functions
def generate_full_report_url(offset):
    return f'https://www.jobmarketinsights.com/jmi-area/FullReport?primaryField=company&scopeId=1120&secondaryField=&period=-1&periodCompare=0&sortTable=market&sortOrder=&offset={offset}&pageSize=100&secondaryOffset=0&category=&company=&location=PH%5EID&title=&jobType=&salary=&comSize=&comType=&hotVacancies=&comJobBoard=&mkt.company=&mkt.site=&time='

def generate_offer_inspect_list_url(formated_company_name):
    return f"https://www.jobmarketinsights.com/jmi-area/OfferInspectList?company=%22{formated_company_name}%22"

def generate_offer_inspect_single_url(offer_id):
    return f"https://www.jobmarketinsights.com/jmi-area/OfferInspectSingle?ids={offer_id}"

inspect_id_pattern = re.compile(r'ids\.push\(\"([\d\^]+)\"\);')
company_encoded_pattern = re.compile(r'jmi_detail\(\'company\', \'(\S+)\'\);\"')
```

```python
headers['Referer'] = 'https://www.jobmarketinsights.com/jmi-area/FullReport?primaryField=company'
```

```python
def get_company_query_names(headers, limit, min_sleep=1., max_sleep=4.):
    company_names = []
    try:
        for offset in tqdm(range(0, limit, 100)):
            url = generate_full_report_url(offset)
            response = requests.post(url, headers=headers)
            headers['Referer'] = url
            company_names += company_encoded_pattern.findall(response.text)
            throttle_request(min_sleep, max_sleep)
    except:
        pass
    
    return company_names

def get_job_ids(company_names, headers, min_sleep=.5, max_sleep=1.):
    summary = OrderedDict()
    try:
        for idx, company_name in tqdm(enumerate(company_names), total=len(company_names)):
            job_ids = []
            url = generate_offer_inspect_list_url(company_name)
            headers['Referer'] = generate_full_report_url(100 * int(idx / 100))
            response = requests.post(url, headers=headers)
            job_ids += inspect_id_pattern.findall(response.text)
            throttle_request(min_sleep, max_sleep)
            summary[company_name] = job_ids
    except Exception as e:
        print(e)
        pass
    
    return summary
```

## Save html files

```python
#n = Top n companies, rand = # of jobs per company, job_ids = set, path = directory 
#Note to self: Make a try... except: continue where u save the list of companies skipped (future debugging)
def save_html(n, rand, job_id_dict, path=''):
    counter = 0
    urls = []
    skipped = []
    top = list(job_id_dict)[:n]
    for t  in tqdm(top):
        jobs = job_ids[t]
        if len(jobs) > rand:
            ids = random.sample(set(jobs), rand)
        else:
            ids = jobs
        for j in range(len(ids)):
            url = generate_offer_inspect_single_url(ids[j])
            #save url in a file
            urls.append(url)
            try:
                response = requests.post(url, headers=headers)
                html = response.text
                html = html.replace('\n', ' ')
                html = html.replace('\t', ' ')
                if path != '':
                    filename = f"{path}/{t}_{j}.html"
                else:
                    filename = f"{t}_{j}.html"
                with io.open(filename, "w", encoding="utf-8") as file:
                    file.write(html)
            except Exception as e:
                print(e)
                skipped.append(url)
                continue
            counter += 1
            throttle_request(0.2, 0.5)
    pd.DataFrame(urls).to_csv('job_urls.csv')
    return skipped, top
```

```python
def backlog(skipped, path=''):
    for url in skipped:
        try:
            response = requests.post(url, headers=headers)
            html = response.text
            html = html.replace('\n', ' ')
            html = html.replace('\t', ' ')
            if path != '':
                filename = f"{path}/{t}_{j}.html"
            else:
                filename = f"{t}_{j}.html"
            with open(filename, "w") as file:
                file.write(html)
        except:
            continue
            
```

## BeautifulSoup parsing

```python
def parse_html(names, rand):
    data = pd.DataFrame(columns=['job_title', 'company', 'location', 'postings'])
    for i in tqdm(names):
        for j in range(rand):
            filename = f"jobs/{i}_{j}.html"
            soup = BeautifulSoup(io.open(filename, encoding="utf-8"), "html.parser")

            try:
                job_title = soup.find_all('h1')[0].get_text()
                company = soup.find_all('h1')[1].get_text()
                location = soup.find_all('span', {'class': 'gray'})[0].get_text()

                #Job Postings
                postings = soup.find_all('li')
                postings_text = []
                for post in postings:
                    site = post.find_all('a')[0].get_text()
                    date = post.contents[2]
                    postings_text.append([date, site])
                postings_dump = json.dumps(postings_text)

                tmp = {
                    'job_title': job_title,
                    'company': company,
                    'location': location,
                    'postings': postings_dump
                }

                tmp = pd.DataFrame(tmp, index=[0])

                data = pd.concat([data, tmp])
            except:
                continue
    data = data.reset_index().drop('index', axis=1)
    return data
```
