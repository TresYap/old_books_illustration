```python
import requests
from bs4 import BeautifulSoup
import time
import random
```

```python
def throttle_request(min_sleep=0.5, max_sleep=1.0):
    time.sleep(random.uniform(min_sleep, max_sleep))
```

```python
#landscape
counter = 0
for i in range(29):
    response = requests.get(f"https://www.oldbookillustrations.com/subjects/landscapes-places/page/{i+1}/")
    soup = BeautifulSoup(response.content, 'html.parser')
    a = soup.find_all("a")
    img_urls = []
    for ref in a:
        if(ref.get_text() == "Open Image Viewer"):
            img_urls.append(ref['href'])

    for img_url in (img_urls):
        filename = f"Landscape\{counter}.jpg"
        try:
            with open(filename, 'wb') as handle:
                response = requests.get(img_url, stream=True)

                if not response.ok:
                    print(response)

                for block in response.iter_content(1024):
                    if not block:
                        break

                    handle.write(block)
        except Exception as e:
            print(e)
            continue
            
        counter += 1
        throttle_request(0.8, 1.5)
```
